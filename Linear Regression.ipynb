```python
import numpy as np
```

```python
#Load data from CSV
data = np.loadtxt("train.csv", delimiter = ",")
data.shape
```

```python
#Split data into X and Y
X = data[:, :13]
X.shape
```

```python
Y = data[:, 13:14]
Y.shape
```

```python
#Split data into testing data and training data
from sklearn import model_selection
X_train, X_test, Y_train, Y_test = model_selection.train_test_split(X,Y)
```

```python
#Import inbuilt linear regression model
from sklearn.linear_model import LinearRegression
alg1 = LinearRegression()
alg1.fit(X_train, Y_train)

print("Value of coefficients: ",alg1.coef_,"Value of intercept", alg1.intercept_)

Y_pred = alg1.predict(X_test)

#Plotting tetsing data with predicted data
import matplotlib.pyplot as plt

one = np.arange(0,60, 1)
plt.plot(one,one)
plt.scatter(Y_test, Y_pred)
plt.show()

#Check score on unknown data
print("Score = ", alg1.score(X_test, Y_test))
```

```python
#Check on training data
Y_pred2 = alg1.predict(X_train)

import matplotlib.pyplot as plt

one = np.arange(0,60, 1)
plt.plot(one,one)
plt.scatter(Y_train, Y_pred2)
plt.show()


print("Score = ",alg1.score(X_train, Y_train))
```
